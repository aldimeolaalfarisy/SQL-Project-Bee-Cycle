```python
# Import required libraries

import pandas as pd
import sqlite3
```

```python
# Import all dataset

df_customer = pd.read_csv('dim_customer.csv')
df_geography = pd.read_csv('dim_geography.csv')
df_product = pd.read_csv('dim_product.csv')
df_territory = pd.read_csv('dim_territory.csv')
df_sales = pd.read_csv('fact_sales.csv')
```

```python
# Create database connection

conn = sqlite3.connect('beecycle_sales.db')
```

```python
# Convert all dataframe to database

df_customer.to_sql('dim_customer', conn, if_exists='replace',index=False)
df_geography.to_sql('dim_geography', conn, if_exists='replace',index=False)
df_product.to_sql('dim_product', conn, if_exists='replace',index=False)
df_territory.to_sql('dim_territory', conn, if_exists='replace',index=False)
df_sales.to_sql('fact_sales', conn, if_exists='replace',index=False)
```

```python
# Close the database

conn.commit()
conn.close()
```
